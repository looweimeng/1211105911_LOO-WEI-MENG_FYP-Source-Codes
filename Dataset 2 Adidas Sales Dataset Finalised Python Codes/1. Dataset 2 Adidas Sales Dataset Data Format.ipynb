# Dataset 2: Adidas Sales Dataset Data Format

## Import Necessary Library

```python
import pandas as pd
import numpy as np
```

## Read Dataset

```python
# Load the Walmart Dataset from Kaggle
df = pd.read_excel('Adidas US Sales Datasets.xlsx')
```

```python
df.head()
```

```python
df = df.drop("Unnamed: 0" , axis=1)
df = df.drop(labels=[0,1,2] , axis=0)
df.columns = df.iloc[0]
df = df.drop(labels=3 , axis=0)
```

```python
df.head()
```

## Save the Dataset into CSV Format

```python
df.to_csv('Adidas.csv', index=False)
print('Adidas.csv has been saved.')
```
